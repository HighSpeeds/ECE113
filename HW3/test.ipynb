```python
import matplotlib.pyplot as plt
import numpy as np

x1=lambda n: [1,-1,1][n+1] if n<=1 and n>=-1 else 0
x2=lambda n:[1,-1,0,1,-1][n] if n<=4 and n>=0 else 0
x3=lambda n:[-1,2,0,-2,1][n+3] if n<=1 and n>=-3 else 0
```

```python
def convolve(f,y,n,r):
    s=0
    for i in r:
        s+=f(i)*y(n-i)
    return s
```

```python
S=[]
for i in range(-10,10):
    S.append(convolve(x1,x1,i,range(-10,10)))
    # S.append(x1(i))

plt.stem(S)
plt.show()
```

```python
S=[]
for i in range(-10,10):
    S.append(convolve(x2,x2,i,range(-10,10)))
    # S.append(x1(i))

plt.stem(range(-10,10),S)
plt.show()
```

```python
S=[]
for i in range(-10,10):
    S.append(convolve(x3,x3,i,range(-10,10)))
    # S.append(x1(i))

plt.stem(range(-10,10),S)
plt.xticks(range(-10,10),range(-10,10))
plt.grid()
plt.show()
```

```python
S=[]

x=lambda n: [3,2,1,1,3,3,2,1,1][n+2] if n<=6 and n>=-2 else 0
y=lambda n: 1 if n>=-1 and n<=1 else 0

for i in range(-10,10):
    S.append(convolve(x,y,i,range(-10,10)))
    # S.append(x1(i))

plt.stem(range(-10,10),S)
plt.xticks(range(-10,10),range(-10,10))
plt.grid()
plt.show()
```
