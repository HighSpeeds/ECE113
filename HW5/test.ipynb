```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
signal=lambda k: 2j*np.sin(2*np.pi/4*k)*(2j*np.cos(np.pi/2*k)+np.exp(2j*np.pi/4*k))
```

```python
s=np.array([signal(k) for k in range(4)])
s
```

```python
fig,axs=plt.subplots(2,1,figsize=(10,5))
plt.sca(axs[0])
plt.stem(range(4),np.abs(s))
plt.ylabel('Magnitude')
plt.sca(axs[1])
angle=np.angle(s)
angle[np.where(np.abs(s)<=1e-1)]=0
plt.stem(range(4),angle)
plt.ylabel('Phase')
plt.xlabel('Frequency Bins')
plt.savefig("fig.png")
plt.show()
```
